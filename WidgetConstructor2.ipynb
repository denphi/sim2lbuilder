```python
%%yaml
inputs:
    float1:
        type : BoundedFloatText
        value : 1.0
        description : Scale
        layout: 
            width : auto
    float2:
        type : FloatText
        value : 1.0        
        description : Angle
        layout: 
            width : auto
    int1:
        type : BoundedIntText
        description : N1
        value : 13
        layout: 
            width : auto
    int2:
        type : IntText    
        value : -7
        description : N2
        layout: 
            width : auto

    int3:
        type : IntText    
        description : N3
        value : -3
        layout: 
            width : auto
    button1:
        type : Button  
        click : ClickButton
        description : Run ClickButton
        layout: 
            width : auto
    button2:
        type : Button  
        click : ClickButton2
        description : Run ClickButton2
        layout: 
            width : auto
outputs:
    out1 : 
        type: FigureWidget
        module: plotly.graph_objects
        layout : 
            title : My 2D plot
    out2 : 
        type: FigureWidget
        module: plotly.graph_objects
        layout : 
            title : My 3D plot
    out3 : 
        type: Output
layout:
    type: HBox
    layout: 
        width : 100%
        height : 600px
        border : 1px solid rgb(198,198,198)
    children:
        container1:
            layout: 
                flex : 1 1
            type: VBox
            children: 
                container11:
                    type: Tab
                    children:
                        container111:
                            type: VBox
                            children:
                                caption: 
                                    type: Label
                                    value : SIN PARAMETERS
                                float1:
                                float2:
                                button1:

                        container112:
                            type: VBox
                            children:
                                caption: 
                                    type: Label
                                    value : SPIROGRAPH PARAMETERS
                                int1:
                                int2:
                                int3:
                                button2:
                    titles : [SIN,SPIROGRAPH]
        container2:
            layout: 
                flex : 2 1
            type: Accordion
            children: 
                out1:
                out2:
                out3:
            titles: [2D, 3D, Output]
```

```python
parameters = _
```

```python
def ClickButton(widget, *kargs):
    import numpy as np
    import plotly.graph_objects as go
    import matplotlib.pyplot as plt
    n1 = widget.inputs["float1"]
    n2 = widget.inputs["float2"]
    out1 = widget.outputs["out1"]
    out2 = widget.outputs["out2"]
    out3 = widget.outputs["out3"]
    t = np.linspace(0, 10, 100)
    y = n1.value * np.sin(n2.value*t)
    out1.data = []
    out1.add_trace(
        go.Scatter(x=t, y=y,
                    mode='markers',
                    name='markers')
    )
    
    out2.data = []
    out2.add_trace(
        go.Scatter3d(x=y, y=t, z=[2 for i in t],
                    mode='markers',
                    name='markers')
    )
    
    out3.clear_output()
    with out3:
        fig = plt.figure()
        plt.plot(t, y)
        plt.show(fig)
        print("This is a plot")
   
def ClickButton2(widget, *kargs):
    import numpy as np
    import plotly.graph_objects as go
    import matplotlib.pyplot as plt
    out1 = widget.outputs["out1"]
    out2 = widget.outputs["out2"]
    out3 = widget.outputs["out3"]
    n1 = widget.inputs["int1"]
    n2 = widget.inputs["int2"]
    n3 = widget.inputs["int3"]
    t = np.linspace(0, 1, 1000)
    z = np.exp(1j*2*np.pi*n1.value*t) + np.exp(1j*2*np.pi*n2.value*t) + np.exp(1j*2*np.pi*n3.value*t)
    t = np.real(z)
    y = np.imag(z)
    
    out1.data = [] 
    out1.add_trace(
        go.Scatter(x=t, y=y,
                    mode='markers',
                    name='markers')
    )
    
    out2.data = []
    out2.add_trace(
        go.Scatter3d(x=y, y=t, z=[2 for i in t],
                    mode='markers',
                    name='markers')
    )
    
    out3.clear_output()
    with out3:
        fig = plt.figure()
        plt.plot(t, y)
        plt.show(fig)
        print("This is a plot")
```

```python
b = WidgetConstructor(parameters)
b.ClickButton = ClickButton
b.ClickButton2 = ClickButton2
b.assemble()
b
```
